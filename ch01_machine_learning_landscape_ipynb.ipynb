## 3) Teori yang Dipakai

### 3.1 Apa itu Machine Learning
Machine Learning (ML) adalah pendekatan untuk membuat sistem yang **belajar pola dari data** sehingga performanya membaik pada tugas tertentu, tanpa perlu menulis aturan (rule) secara eksplisit untuk setiap kasus.

**Kapan ML lebih cocok daripada rule-based:**
- Aturan terlalu banyak/rumit untuk ditulis manual (mis. filter spam, rekomendasi, pengenalan gambar).
- Pola berubah seiring waktu (mis. fraud, tren pengguna) sehingga aturan cepat usang.
- Data tersedia cukup, dan ada metrik yang jelas untuk mengevaluasi performa.

### 3.2 Jenis-jenis Sistem ML (berdasarkan data/tujuan)
- **Supervised Learning**: data punya label (y). Contoh: klasifikasi (spam/tidak), regresi (prediksi harga).
- **Unsupervised Learning**: data tanpa label. Contoh: clustering, dimensionality reduction, anomaly detection.
- **Semi-supervised Learning**: sebagian kecil data berlabel, sisanya tidak.
- **Self-supervised Learning**: label “dibuat” dari data itu sendiri (mis. memprediksi bagian yang hilang).
- **Reinforcement Learning (RL)**: agen belajar dari reward/penalti melalui interaksi dengan environment.

### 3.3 Cara Belajar: Batch vs Online
- **Batch learning**: model dilatih dari kumpulan data (batch) secara offline. Update model dilakukan periodik.
  - Kelebihan: stabil, mudah direproduksi.
  - Kekurangan: tidak cepat beradaptasi jika data berubah.
- **Online learning**: model diupdate bertahap dari data yang datang (stream).
  - Kelebihan: adaptif, cocok untuk data besar/streaming.
  - Kekurangan: sensitif terhadap data buruk (noise) dan butuh monitoring.

### 3.4 Generalisasi: Train/Validation/Test
Tujuan ML bukan hanya “hafal” data training, tetapi **mengeneralisasi** ke data baru.
- **Train set**: untuk melatih model.
- **Validation set**: untuk memilih model/hyperparameter.
- **Test set**: evaluasi akhir (tidak dipakai saat tuning).

### 3.5 Overfitting vs Underfitting
- **Overfitting**: model terlalu kompleks → performa bagus di train, buruk di data baru.
  - Solusi umum: tambah data, regularisasi, kurangi kompleksitas, early stopping, cross-validation.
- **Underfitting**: model terlalu sederhana → buruk bahkan di train.
  - Solusi umum: gunakan model lebih kompleks, fitur lebih informatif, kurangi regularisasi.

### 3.6 Istilah Dasar yang Harus Paham
- **Feature (X)**: input/atribut.
- **Label/Target (y)**: output yang ingin diprediksi.
- **Model**: fungsi pemetaan X → y.
- **Parameter**: nilai yang dipelajari saat training (mis. bobot).
- **Hyperparameter**: pengaturan yang ditentukan sebelum training (mis. learning rate, depth, C).
- **Metric**: ukuran performa (accuracy, precision/recall, RMSE, dll).

### 3.7 Gambaran Pipeline ML
Kumpulan langkah umum:
1) definisikan masalah + metrik  
2) kumpulkan & pahami data  
3) preprocessing & feature engineering  
4) training + validasi  
5) evaluasi test  
6) deploy  
7) monitoring (drift/performa) + retraining bila perlu


## 4) Takeaways
- ML dipakai saat aturan manual terlalu kompleks/rapuh, atau pola berubah dari waktu ke waktu.
- Klasifikasi sistem ML penting: supervised vs unsupervised, batch vs online, serta peran RL.
- Fokus utama ML adalah **generalisasi**, bukan sekadar performa di data training.
- Overfitting dan underfitting adalah masalah inti; pemilihan model, data, dan regularisasi jadi kuncinya.
